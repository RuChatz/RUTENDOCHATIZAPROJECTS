# Classifying Iris Species Exercise 

This is a classification problem in which a machine learning model is used to catergorise a new iris species into one of the three iris species found in the dataset namely Setosa , Versicolor and Virginica . The model uses a k-Nearest Neighbors classification algorithm to sort each iris variant .
This is a supervised learning problem as there is input data ( various measurements of the parts of the flowers ) and the expected output data i.e the corresponding species .

# 1 . Understanding the data 

In the first portion of the code I seek to understand the scope of the data .

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import load_iris

iris_dataset = load_iris()
print('Keys of iris_dataset : \n{}'.format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193]+'\n...')
```

```python
print('Feature names : \n{}'.format(iris_dataset['feature_names']))
print('Type of data : \n{}'.format(type(iris_dataset['data'])))
print('Shape of data : {}'.format(iris_dataset['data'].shape))
```

```python
#The first five samples' feature values :
print('First five columns of data : \n{}'.format(iris_dataset['data'][:5]))
print('Type of target :{}'.format(type(iris_dataset['target'])))
print('Shape of target : {}' .format(iris_dataset['target'].shape))
print('Target : \n{}'.format(iris_dataset['target']))
#iris['target_names'] array : 0 means setosa,  1 means versicolor and 2 means virginica 
```

#  2. Measuring Success : Training and Testing Data 

In this portion I aim to partition the data into training data and testing data that will be fit into the machine learning model .The training data is used to build our model while the testing data is used to assess the functionality of the model.

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
print('X_train shape : {}'.format(X_train.shape))
print('y_train shape : {}'.format(y_train.shape))
print('X_test shape : {}'.format(X_test.shape))
print('y_test shape : {}'.format(y_test.shape))
```

# 3.Visualising the Data 



```python
#creating a dataframe from Data in X_train 
iris_df = pd.DataFrame(X_train , columns=iris_dataset.feature_names)

#creating a scatter matrix from the dataframe , color by y_train

grr= pd.plotting.scatter_matrix(iris_df,c=y_train , figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```

# 4.Choosing the appropriate model 

k-Nearest Neighbors (k-NN) is a non-parametric , supervised learning classifier , which uses proximity to make classifications or predictions about the grouping of an individual data point . 

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train , y_train)
```

# 5. Predictions

```python
#Adding new measurements of a new flower into a row in a 2D Numpy array

X_new = np.array([[5,2.9,1,0.2]])
print('X_new shape : {}'.format(X_new.shape))
prediction = knn.predict(X_new)
print('Prediction : {}'.format(prediction))
print('Predicted target name : {}'.format(iris_dataset['target_names'][prediction]))
```

# 6. Evaluation of the Model

This portion checks the accuracy of the model so as to judge whether this model is the right fit for this task or not .

```python
y_pred = knn.predict(X_test)
print('Test set prediction : \n{}'.format(y_pred))
print('Test set score : {:.2f}'.format(np.mean(y_pred == y_test)))
print('Test set score : {:.2f}'.format(knn.score(X_test , y_test)))
```

The model is 97% accurate making it a suitable fit for the task .
